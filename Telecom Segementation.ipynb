```python
# dropping the variables
df.drop(['equipmon','wiremon'],axis=1,inplace=True)
```

```python
# checking for the outliers

def var_summary(x):
    return pd.Series([x.count(), x.isnull().sum(), x.sum(), x.mean(), x.median(),  x.std(), x.var(), x.min(), x.dropna().quantile(0.01), x.dropna().quantile(0.05),x.dropna().quantile(0.10),x.dropna().quantile(0.25),x.dropna().quantile(0.50),x.dropna().quantile(0.75), x.dropna().quantile(0.90),x.dropna().quantile(0.95), x.dropna().quantile(0.99),x.max()], 
                  index=['N', 'NMISS', 'SUM', 'MEAN','MEDIAN', 'STD', 'VAR', 'MIN', 'P1' , 'P5' ,'P10' ,'P25' ,'P50' ,'P75' ,'P90' ,'P95' ,'P99' ,'MAX'])
num_summary = df.apply(lambda x:var_summary(x)).T
```

```python
num_summary
```

```python
#removing the outliers

def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x
df = df.apply(lambda x:outlier_capping(x))
```

```python
num_summary = df.apply(lambda x: var_summary(x)).T
```

```python
num_summary
```

```python
#  Standarize the data

from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
df_scaled = sc.fit_transform(df)
```

```python
pd.DataFrame(df_scaled).describe()
```

```python
# applying the PCA
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components=28)
pc.fit(df_scaled)
```

```python
pc.explained_variance_
```

```python
var1=np.cumsum(np.round(pc.explained_variance_ratio_, decimals=4)*100)
```

```python
var1
```

```python
plt.plot(var1)
```

```python
pc_final = PCA(n_components=18).fit(df_scaled)
final_df = pc_final.fit_transform(df_scaled)
```

```python
pd.DataFrame(final_df)
```

```python
from sklearn.cluster import KMeans
```

```python
km_3 = KMeans(n_clusters=3,random_state=123)
km_3.fit(final_df)
```

```python
km_4=KMeans(n_clusters=4,random_state=123).fit(final_df)
#km_5.labels_a

km_5=KMeans(n_clusters=5,random_state=123).fit(final_df)
#km_5.labels_

km_6=KMeans(n_clusters=6,random_state=123).fit(final_df)
#km_6.labels_

km_7=KMeans(n_clusters=7,random_state=123).fit(final_df)
#km_7.labels_

km_8=KMeans(n_clusters=8,random_state=123).fit(final_df)
#km_5.labels_
```

```python
pd.Series(km_3.labels_).value_counts()
```

```python
df['cluster_3'] = km_3.labels_
df['cluster_4'] = km_4.labels_
df['cluster_5'] = km_5.labels_
df['cluster_6'] = km_6.labels_
df['cluster_7'] = km_7.labels_
df['cluster_8'] = km_8.labels_
```

```python
df.head()
```

```python
pd.Series.sort_index((df.cluster_3).value_counts())
```

```python
pd.Series(df.cluster_3.size)
```

```python
size=pd.concat([pd.Series(df.cluster_3.size), pd.Series.sort_index(df.cluster_3.value_counts()), pd.Series.sort_index(df.cluster_4.value_counts()),
           pd.Series.sort_index(df.cluster_5.value_counts()), pd.Series.sort_index(df.cluster_6.value_counts()),
           pd.Series.sort_index(df.cluster_7.value_counts()), pd.Series.sort_index(df.cluster_8.value_counts())])
```

```python
Seg_size=pd.DataFrame(size, columns=['Seg_size'])
Seg_Pct = pd.DataFrame(size/df.cluster_3.size, columns=['Seg_Pct'])
Seg_size.T
```

```python
Profling_output = pd.concat([df.apply(lambda x: x.mean()).T, df.groupby('cluster_3').apply(lambda x: x.mean()).T, df.groupby('cluster_4').apply(lambda x: x.mean()).T,
          df.groupby('cluster_5').apply(lambda x: x.mean()).T, df.groupby('cluster_6').apply(lambda x: x.mean()).T,
          df.groupby('cluster_7').apply(lambda x: x.mean()).T, df.groupby('cluster_8').apply(lambda x: x.mean()).T], axis=1)

Profling_output_final=pd.concat([Seg_size.T, Seg_Pct.T, Profling_output], axis=0)

Profling_output_final.columns = ['Overall', 'KM3_1', 'KM3_2', 'KM3_3',
                                'KM4_1', 'KM4_2', 'KM4_3', 'KM4_4',
                               'KM5_1', 'KM5_2', 'KM5_3', 'KM5_4', 'KM5_5',
                                'KM6_1', 'KM6_2', 'KM6_3', 'KM6_4', 'KM6_5','KM6_6',
                                'KM7_1', 'KM7_2', 'KM7_3', 'KM7_4', 'KM7_5','KM7_6','KM7_7',
                                'KM8_1', 'KM8_2', 'KM8_3', 'KM8_4', 'KM8_5','KM8_6','KM8_7','KM8_8',]
```

```python
# Selecting the optimal number of cluster
from sklearn import metrics
# calculate SC for K=3 through K=12
k_range = range(3, 12)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(final_df)
    scores.append(metrics.silhouette_score(final_df, km.labels_))
```
